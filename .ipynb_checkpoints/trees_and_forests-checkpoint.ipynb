## Implementacija potrebnih funkcija za učitavanje podataka

```python
def get_classes_info(y: np.ndarray, show_text: bool=True) -> list:
    
    """
        Funkcija za izvlacenje klasa i prebrojavanja primera
        sa istom klasom
        
        params: 
            - y: vektor klasa
                               
        :return:
            - classes: lista klasa
            - classes_dict: recnik klasa sa njihovim brojem predstavnika
    """
    # dohvatanje svih vrednosti klasa
    classes = list(set(y.astype('int')))
    
    assert classes == [0,1], 'Klase nisu odgovarajucih vrednosti!'
            
    # prebrojavanje predstavnika klasa
    classes_dict = dict.fromkeys(classes)
    for _class in classes:
        class_count = np.count_nonzero(y == _class)
        classes_dict[_class] = class_count
        
    # broj klasa
    k = len(classes_dict.keys())
        
    # ispis podataka o klasama u dataset-u
    if show_text:
        print('-------------------------------------------------------')
        print(f'Ukupan broj klasa kojima pripadaju ucitani primeri je: {k}\n')
        print('Ucitane klase i brojevi njhovih predstavnika/primera su: ')
        for _class in classes_dict.keys():
            pr = 'primera'
            if classes_dict[_class] % 10 == 1:
                pr = 'primer'
            print(f'\tKlasa {_class}: {classes_dict[_class]} {pr}')
        print('-------------------------------------------------------')
    
    return classes, classes_dict
```

```python
def read_data(path: str='data_1.csv', separator_index: int=-1) -> tuple:
    
    """
        Funkcija za ucitavanje podataka na osnovu kojih se
        kreira model.
        
        :params: 
            - path: putanja do fajla sa podacima
            - separator_index: broj kolone do koje se nalaze
                               ulazni podaci, odnosno od koje
                               pocinju izlazni podaci
                               
        :return:
            - X: matrica primera svih prediktora
            - y: vektor izlaza
    """
    
    data = pd.read_csv(path).to_numpy()
    
    # razdvajanje ulaznih i izlaznih promenljivih
    X, y = data[:,0:separator_index], data[:,separator_index]
    
    return X, y
```

## Implementacija funkcija za podelu podataka na trening i validacioni skup

```python
def train_test_data_split(X: np.ndarray, y: np.ndarray, split_ratio: float=0.8) -> tuple:
    
    """
        Funkcija za mesanje klasa i splitovanje svih dostupnih podataka na 
        trening/obucavajuce i test podatke prema zadatom postotku (split_ratio).
        U funkciji se vodi racuna da balans izmedju klasa u dva nova skupa podataka
        bude isti kao i kod originalnih prosledjenih podataka.
        
        :params:
            - X: matrica primera svih prediktora
            - y: vektor klasa 
            - split_ratio: vrednost iz intervala [0,1] koja oznacava procenat
                           trening podataka
                           
        :return:
            - X_train: matrica primera svih prediktora za trening skup
            - y_train: vektor izlaza za trening skup
            - X_test: matrica primera svih prediktora za test skup
            - y_test: vektro izlaza za test skup
    """
    
    random.seed(1)
    
    # broj primera trening skupa
    m_train = 0
    
    classes, classes_dict = get_classes_info(y, False)
    
    train_indices = set()
    test_indices = set()
    
    # formiranje zasebnih splitova, kako bi svaka klasa imala split_ratio * 100% odbiraka u trening skupu
    for i in classes:
        m = classes_dict[i]
        
        class_train_set_size = int(m*split_ratio)
        m_train += class_train_set_size

        class_ind = np.where(y == i)[0].tolist()

        # odredjivanje indeksa za trening skup za jednu klasu
        class_train_set_ind = random.sample(class_ind, k=class_train_set_size)
        
        # dodavanje u globalni set indeksa za trening skup
        train_indices.update(list(class_train_set_ind))

        # odredjivanje indeksa za test skup za jednu klasu
        class_test_set_ind = set(class_ind) - set(class_train_set_ind)
        
        # dodavanje u globalni set indeksa za test skup
        test_indices.update(list(class_test_set_ind))
        
    train_indices = list(train_indices)
    test_indices = list(test_indices)
    
    # podela podataka na trening i test podatke
    X_train, y_train = X[train_indices], y[train_indices]
    X_test, y_test = X[test_indices], y[test_indices]
    
    # mesanje podataka
    X_train, y_train = shuffle(X_train, y_train, random_state=1)
    X_test, y_test = shuffle(X_test, y_test, random_state=1)
    
    return X_train, y_train, X_test, y_test
```

## Implementacija funkcija za sortiranje prediktora

```python
def predictors_correlation(X: np.ndarray, y: np.ndarray) -> list:
    
    """
        Funkcija za racunanje koeficijenta korelacije prediktora 
        sa izlazom.
        
        :params:
            - X: matrica primera svih prediktora
            - y: vektor klasa
                           
        :return:
            - predictor_indices: lista indeksa sortiranih prediktora
    """

    y = y.flatten()
    
    # odredivanje korelacije prediktora iz izlaza
    corr_coef = [abs(np.corrcoef(X[:,i], y)[0,1]) for i in range(X.shape[1])]
    
    # dohvatanje u listu indeksa prediktora
    predictor_indices = list(np.arange(X.shape[1]))

    # sortiranje
    corr_coef, predictor_indices = (list(t) for t in zip(*sorted(zip(corr_coef, predictor_indices))))

    # sortiranje od najboljeg ka najgorem
    corr_coef.reverse()
    predictor_indices.reverse()

    return predictor_indices
```

```python
def get_folds(X: np.ndarray, y: np.ndarray, k: int=5) -> dict:
    
    """
        Funkcija za dohvatanje svih foldova za treniranje i validaciju.
        
        :params:
            - X: matrica primera svih prediktora
            - y: vektor izlaza
            - k: ukupan broj foldova
            
        :return:
            - folds: recnik koji sadzi sve podatke za sve foldove
    """

    # kreiranje recnika za cuvanje podataka
    folds = {}
    
    # formiranje liste svih mogucih indeksa
    m = X.shape[0]
    indices = list(np.arange(m))
    
    # kreiranje skupa indeksa iskoriscenih za validacioni skup
    used_val_fold_indices = set()
    
    # prolazak po petlji za svaki fold
    for i in range(k):
        
        # dohvatanje indeksa za validacioni skup
        val_fold_indices = get_val_fold_indices(y, m, k, used_val_fold_indices)
        
        # dohvatanje indeksa za trening skup
        train_fold_indices = list(set(indices) - set(val_fold_indices))
        
        # uzorkovanje validacionih podataka
        X_val, y_val = X[val_fold_indices, :], y[val_fold_indices]
        
        # uzorkovanje trening podataka
        X_train, y_train = X[train_fold_indices, :], y[train_fold_indices]
        
        # cuvanje podataka u recnik
        folds[i] = [(X_val, y_val), (X_train, y_train)] 
        
        # dodavanje iskoriscenih indeksa u skup
        used_val_fold_indices = used_val_fold_indices.union(set(val_fold_indices))
        
    return folds
```

```python
def get_val_fold_indices(y: np.ndarray, m: int, k: int, used_val_fold_indices: set) -> list:
    
    """
        Funkcija za dohvatanje indeksa za validacionih skup.
        
        :params:
            - y: vektor klasa
            - m: broj primera celog skupa podataka
            - k: broj foldova
            - used_val_fold_indices: skup vec iskoriscenih indeksa
            
        :return:
            - val_fold_indices: lista indeksa za validacioni skup
    """
    
    random.seed(1)
    
    # odredjivanje velicine folda
    fold_size = int(m/k)
    
    # dohvatanje svih indeksa
    indices = list(np.arange(m))
    
    # odredjivanje svih mogucih indeksa naspram svih vec iskoriscenih
    available_indices = set(indices) - used_val_fold_indices
    
    # odredjivanje svih mogucih indeksa za 0. i 1. klasu
    available_indices_0 = available_indices.intersection(set(np.squeeze(np.argwhere(y==0))))
    available_indices_1 = available_indices.intersection(set(np.squeeze(np.argwhere(y==1))))
    
    # odredjivanje velicine skupa svih mogucih indeksa za 0. i 1. klasu
    len_0 = len(available_indices_0)
    len_1 = len(available_indices_1)
    
    # odredjivanje velicine jedne klase u okviru jednog folda
    fold_size_0 = int(len_0/(len_0 + len_1) * fold_size)
    fold_size_1 = int(len_1/(len_0 + len_1) * fold_size)
    
    # uzorkovanje indeksa za svaku klasu
    fold_0 = random.sample(available_indices_0, k=fold_size_0)
    fold_1 = random.sample(available_indices_1, k=fold_size_1)
    
    # sakupljanje svih indeksa
    val_fold_indices = fold_0 + fold_1
    
    return val_fold_indices
```

```python
def standardize(X: np.ndarray, mu: np.ndarray=None, std: np.ndarray=None) -> np.ndarray:
    
    """
        Funkcija za standardizaciju prediktora - postavljanje prediktora na istu skalu.
        Svaka kolona matrice X se standardizuje sa sebi svojstvenim statistikama - mu i std,
        s obzirom na to da su to posebni prediktori i da u opstem slucaju ne moraju biti na istoj skali.
        
        :params:
            - X: matrica primera svih prediktora
            - mu: vektor srednjih vrednosti za svaki prediktor ili None
            - std: vektor standardnih devijacija za svaki prediktor ili None
            
        :return:
            - X: standardizovana matrica primera svih prediktora
            - mu: vektor srednjih vrednosti za svaki prediktor
            - std: vektor standardnih devijacija za svaki prediktor
    """
    
    # racunanje srednjih vrednosti i st. devijacije svih prediktora
    if mu is None and std is None:
        mu, std = np.mean(X, axis=0), np.std(X, axis=0)

    # standardizacija na normalnu raspodelu
    X = (X - mu) / std # ~ N(0,1)
    
    return X, mu, std
```

```python
def cross_validation(X: np.ndarray, y: np.ndarray, k: int = 5) -> tuple:
    
    """
        Funkcija za izvrsavanje unakrsne validacije - potrebne za "omotac algoritam",
        sa logistickom regresijom bez regularizacije.
        
        :params:
            - X: matrica primera svih prediktora
            - y: vektor klasa
            - k: broj foldova preko kojih se radi unakrsna validacija
                           
        :return:
            - train_mean_acc: usrednjena tacnost modela na trening skupu
            - val_mean_acc: usrednjena tacnost modela na validacionom skupu
    """
    
    # dohvatanje foldova
    folds = get_folds(X, y, k)
    
    # inicijalizacija listi za cuvanje vrednosti metrika
    acc_train, acc_val = [], []
    
    # petlja po iteracijama unakrsne validacije
    for i in range(k):
        # dohvatanje validacionog i trening skupa
        val, train = folds[i]

        # otpakivanje skupova na ulazne i izlazne 
        X_train, y_train = train[0], train[1]
        X_val, y_val = val[0], val[1]
        
        y_train = y_train.flatten()
        y_val = y_val.flatten()
        
        # standardizacija prediktora trening skupa
        X_train, mu_train, std_train = standardize(X_train)

        # standardizacija prediktora validacionog skupa sa 
        # statistikama iz trening skupa
        X_val, _, _ = standardize(X_val, mu_train, std_train)
        
        # kreiranje modela - logisticka regresija
        model = LogisticRegression(penalty='none', random_state=1)
        
        # obucavanje modela nad trening podacima
        model.fit(X_train, y_train)
        
        # racunanje predikcija modela
        y_train_hat = model.predict(X_train)
        y_val_hat = model.predict(X_val)
        
        # racunanje tacnosti dobijenih predikcija
        train_acc = (sum(y_train.flatten() == y_train_hat.flatten()) / y_train.size)
        val_acc = (sum(y_val.flatten() == y_val_hat.flatten()) / y_val.size)
        
        # Cuvanje vrednosti na trenutnom struku
        acc_train.append(train_acc)
        acc_val.append(val_acc)
        
    acc_train, acc_val = np.array(acc_train), np.array(acc_val)

    # srednja vrednost tacnosti
    train_mean_acc, val_mean_acc = np.mean(acc_train), np.mean(acc_val)
    
    # standardna devijacija tacnosti
    train_std_acc, val_std_acc = np.std(acc_train), np.std(acc_val)

    return train_mean_acc, val_mean_acc
```

```python
def append_predictor(pred: np.ndarray, i: int) -> np.ndarray:

    """
        Funkcija za dodavanje prediktora u niz.

        :params:
            - pred: 1D niz sa bool vrednostima za odgovarajuce prediktore
            - i: indeks prediktora kojeg treba dodati

        :return:
            - x: novi niz prediktora
    """

    x = deepcopy(pred)
    x[i] = True
    
    return x
```

```python
def predictors_wrapper(X: np.ndarray, y: np.ndarray, k: int = 5) -> list:
    
    """
        Funkcija za izvrsavanje "omotac" algoritma.

        :params:
            - X: matrica primera svih prediktora
            - y: vektor klasa
            - k: broj foldova preko kojih se radi unakrsna validacija

        :return:
            - x: novi niz prediktora
    """

    # Niz iskoriscenih prediktora
    used = np.zeros(X.shape[1], dtype=bool)
    # List sortiranih prediktora
    pred_sorted = []

    while np.sum(used) != used.size:

        # Kreiranje novih prediktora
        feature_combinations = [append_predictor(used, i) for i in range(X.shape[1]) if not used[i]]
    
        # Kros validacija sa novim prediktorima
        res = list(map(lambda x: cross_validation(X[:,x], y, k), feature_combinations))

        # Tacnosti na obucavajucem i validacionom skupu
        train_acc, val_acc = list(map(lambda x: x[0], res)), list(map(lambda x: x[1], res))

        # Prediktor sa najvecom tacnoscu na validacionom skupu
        ind_best = np.argmax(val_acc)
        best_combination = feature_combinations[ind_best]
        best_pred = np.argmin(best_combination == used)
        pred_sorted.append(best_pred)

        # Prepisivanje iskoriscenih prediktora
        used = best_combination

    return pred_sorted
```

## Implementacija funkcije za treniranje modela (Logisticke regresije) sa podskupovima prediktora

```python
def train_with_predictors_subset(X_train: np.ndarray, y_train: np.ndarray, X_test: np.ndarray, y_test: np.ndarray, sorted_predictors: list, method: str) -> None:
    
    """
        Funkcija za treniranje modela - logisticke regresije nad trening i test skupom podataka,
        sa podskupovima prediktora.

        :params:
            - X_train: matrica primera svih prediktora iz trening skupa
            - y_train: vektor klasa iz trening skupa
            - X_test: matrica primera svih prediktora iz fiksiranog validacionog (test) skupa
            - y_test: vektor klasa iz fiksiranog validacionog (test) skupa
            - sorted_predictors: lista sortiranih prediktora, od najboljeg kao najlosijem
            - method: metod kojim je dobijen redosled prediktora, sluzi za naslov grafika

        :return:
            - None
    """

    # alokacija nizz trenutno koriscenih prediktora
    predictors = np.zeros(X_train.shape[1], dtype=bool)
    
    # standardizacija prediktora trening skupa
    X_train, mu_train, std_train = standardize(X_train)

    # standardizacija prediktora test skupa sa 
    # statistikama iz trening skupa
    X_test, _, _ = standardize(X_test, mu_train, std_train)
    
    y_train, y_test = y_train.flatten(), y_test.flatten()

    # kreiranje listi za cuvanje tacnosti
    train_acc, test_acc = [], []

    # inkrementalna petlja po prediktorima
    for predictor in sorted_predictors:
        
        # dodavanje prediktora
        predictors[predictor] = True

        # izdvajanje primera selektovanih prediktora po skupovima podataka
        X_train_, X_test_ = X_train[:, predictors], X_test[:, predictors]

        # kreiranje modela - logisiticke regresije
        model = LogisticRegression(penalty='none', random_state=1)
        
        # obucavanje modela
        model.fit(X_train_, y_train)
        
        # izvrsavanje predikcije obucenog modela
        y_train_hat = model.predict(X_train_)
        y_test_hat = model.predict(X_test_)
        
        # racunanje tacnosti dobijenih predikcija
        train_acc_ = (sum(y_train.flatten() == y_train_hat.flatten()) / y_train.size) 
        test_acc_ = (sum(y_test.flatten() == y_test_hat.flatten()) / y_test.size) 
        
        train_acc.append(train_acc_)
        test_acc.append(test_acc_)

    # Prikaz rezultata
    x_axis = np.arange(X_train.shape[1])
    fig = plt.figure(figsize=(18,8))
    plt.plot(x_axis, train_acc)
    plt.plot(x_axis, test_acc)
    plt.xticks(ticks=x_axis, labels=sorted_predictors)
    plt.legend(['Obucavajuci skup','Validacioni/Test skup'], loc='lower right')
    plt.title(f'Zavisnost tacnosti modela LogisticRegression od izabranih prediktora \ndobijenih metodom - {method}')
    plt.ylabel('Tacnost')
    plt.grid()
    plt.show()
```

## Implementacija funkcija za obučavanje stabla

```python
def train_decision_tree_with_variable_depth(X_train: np.ndarray, y_train: np.ndarray, X_test: np.ndarray, y_test: np.ndarray, depth_list: list, display: bool=True) -> None:
    
    """
        Funkcija za treniranje DecisionTreeClassifier modela sa razlicitim maksimalnim dubinama stabla i 
        iscrtavanje zavisnosti greske od dubine stabla
        
        :params:
            - X_train: matrica primera svih prediktora na trening skupu
            - y_train: vektor klasa na trening skupu
            - X_test: matrica primera svih prediktora na validacionom/test skupu
            - y_test: vektor klasa na validacionom/test skupu
            - depth_list: lista sa maksimalnim dubinama stabla
            - display: bool indikator za plotovanje
                           
        :return:
            - None
    """
    
    # inicijalizacija listi za pracenje greski na trening i validacionom/test skupu
    train_error_cnt, test_error_cnt = [], []
    
    for depth in depth_list:
        # kreiranje modela - decision tree
        model = DecisionTreeClassifier(max_depth=depth, random_state=1)

        # fitovanje modela
        model.fit(X_train, y_train)

        # odredjivanje predikcije modela nad istreniranim modelom
        y_train_hat = model.predict(X_train)
        y_test_hat = model.predict(X_test)

        # odredjivanje broja pogresno klasifikovanih primera na trening i validacionom/test skupu
        train_err = sum(y_train.flatten() != y_train_hat.flatten())
        test_err = sum(y_test.flatten() != y_test_hat.flatten())
        
        train_error_cnt.append(train_err)
        test_error_cnt.append(test_err)
    
    # iscrtavanje rezultata
    if display:
        fig = plt.figure(figsize=(18,8))
        plt.plot(depth_list, train_error_cnt, label='Obucavajuci skup')
        plt.plot(depth_list, test_error_cnt, label='Validacioni skup')
        plt.plot(depth_list, np.array(train_error_cnt) + np.array(test_error_cnt), label='Ukupno')
        plt.title('Zavisnost broja pogrešno klasifikovanih primera od maksimalne dubine stabla')
        plt.ylabel('# pogrešno klasifikovanih primera')
        plt.xlabel('Maksimalna dubina stabla')
        plt.legend(loc='best')
        plt.grid()
        plt.show()
```

```python
def plot_decision_boundary(clf: object, X_train: np.ndarray, y_train: np.ndarray, X_test: np.ndarray, y_test: np.ndarray, title: str) -> None:
    
    """
        Funkcija za iscrtavanje podataka i separacionih oblasti.
        
        :params:
            - clf: klasifikator objekat
            - X_train: matrica primera svih prediktora na trening skupu
            - y_train: vektor klasa na trening skupu
            - X_test: matrica primera svih prediktora na validacionom/test skupu
            - y_test: vektor klasa na validacionom/test skupu
            - title: naslov plot-a
                           
        :return:
            - None
    """
    
    # odredjivanje minimuma/maksimuma po obe ose za plot
    xmin, xmax = min(X_train[:,0].min(), X_test[:,0].min()) - 1, max(X_train[:,0].max(), X_test[:,0].max()) + 1
    ymin, ymax = min(X_train[:,1].min(), X_test[:,1].min()) - 1, max(X_train[:,1].max(), X_test[:,1].max()) + 1
    
    xx, yy = np.meshgrid(
      np.linspace(xmin, xmax, num=100, endpoint=True), 
      np.linspace(ymin, ymax, num=100, endpoint=True))
    
    # odredjivanje predikcije modela
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # iscrtavanje
    fig, ax = plt.subplots(ncols=1, figsize=(12,10))
    cs = plt.contourf(xx, yy, Z, alpha=0.2)
    ax.set_title(title)
    plt.scatter(X_train[y_train==1, 0], X_train[y_train==1, 1], s = 50, c = 'blue', marker='o', label = 'Obucavajuci skup, y = 1')
    plt.scatter(X_test[y_test==1, 0], X_test[y_test==1, 1], s = 50, c = 'blue', marker='*', label = 'Validacioni/Test skup, y = 1')
    plt.scatter(X_train[y_train==0, 0], X_train[y_train==0, 1], s = 50, c = 'darkred', marker='o', label = 'Obucavajuci skup, y = 0')
    plt.scatter(X_test[y_test==0, 0], X_test[y_test==0, 1], s = 50, c = 'darkred', marker='*', label = 'Validacioni/Test skup, y = 0')
    ax.set_xlabel("$X_{1}$")
    ax.set_ylabel("$X_{2}$")
    plt.legend(loc='best')
    plt.grid()
    plt.show()
```

```python
def train_and_plot_decision(X_train: np.ndarray, y_train: np.ndarray, X_test: np.ndarray, y_test: np.ndarray, depth_dict: dict) -> None:
    
    """
        Funkcija za treniranje DecisionTreeClassifier modela sa razlicitim maksimalnim dubinama stabla i 
        iscrtavanje podataka sa granicom odluke.
        
        :params:
            - X_train: matrica primera svih prediktora na trening skupu
            - y_train: vektor klasa na trening skupu
            - X_test: matrica primera svih prediktora na validacionom/test skupu
            - y_test: vektor klasa na validacionom/test skupu
            - depth_dict: recnik sa dubinama i naslovima plotova
                           
        :return:
            - None
    """
    
    for depth in depth_dict.keys():
        # kreiranje modela - decision tree
        model = DecisionTreeClassifier(max_depth=depth, random_state=1)

        # fitovanje modela
        model.fit(X_train, y_train)
        
        # iscrtavanje odbiraka i granice 
        plot_decision_boundary(model, X_train, y_train, X_test, y_test, depth_dict[depth])
```

# Ansambli

## Implementacija funkcija za ansamble metode

```python
def train_ensemble_method(clf: str, X_train: np.ndarray, y_train: np.ndarray, X_test: np.ndarray, y_test: np.ndarray,
                          n_estimators_list: list, param_type: str, params_list: list, display: bool=True) -> None:
    
    """
        Funkcija za treniranje RandomForestClassifier modela sa vrednostima jednog hiper-parametra.
        
        :params:
            - clf: string iz skupa {'RF', 'GB'} koji oznacava koji je klasifikator u pitanju
            - X_train: matrica primera svih prediktora na trening skupu
            - y_train: vektor klasa na trening skupu
            - X_test: matrica primera svih prediktora na validacionom/test skupu
            - y_test: vektor klasa na validacionom/test skupu
            - n_estimators_list: lista sa velicinama ansambala
            - param_type: vrsta hiper-parametra
            - params_list: vrednosti hiper-parametra
            - display: bool indikator za prikaz
                           
        :return:
            - None
    """
    
    # kreiranje recnika za cuvanje tacnosti
    train_acc, test_acc = dict.fromkeys(params_list), dict.fromkeys(params_list)
    
    # petlja po svim vrednostima parametara
    for param in params_list:
        
        train_acc[param] = []
        test_acc[param] = []
        
        # petlja po svim vrednostima velicine ansambla
        for n_estimators in n_estimators_list:

            # kreiranje modela 
            if clf.upper() == 'RF':
                model = RandomForestClassifier(n_estimators=n_estimators, random_state=1, **({param_type: param}))
                
            elif clf.upper() == 'GB':
                model = GradientBoostingClassifier(n_estimators=n_estimators, random_state=1, **({param_type: param}))

            y_train = y_train.flatten()
            y_test = y_test.flatten()
            
            # obucavanje modela
            model.fit(X_train, y_train)

            # predikcija modela
            y_train_hat = model.predict(X_train)
            y_test_hat = model.predict(X_test)

            # racunanje tacnosti modela
            train_acc_ = (sum(y_train == y_train_hat.flatten())/y_train.size)
            test_acc_ = (sum(y_test == y_test_hat.flatten())/y_test.size)

            train_acc[param].append(train_acc_)
            test_acc[param].append(test_acc_)
        
    # iscrtavanje grafika
    if display:
        fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(16,14))
        ax = axes.ravel()
        title_string = 'RandomForest' if clf.upper() == 'RF' else 'GradientBoosting'
        fig.suptitle(f'Zavisnost tacnosti {title_string} algoritma od hiper-parametra "{param_type}"')
        
        for param in params_list:
            ax[0].plot(n_estimators_list, train_acc[param], label=f'{param_type} = {param}')
            ax[1].plot(n_estimators_list, test_acc[param], label=f'{param_type} = {param}')
            
        ax[0].set_title('Obucavajuci skup')
        ax[0].set_xlabel('Velicina ansambla')
        ax[0].set_ylabel('Tacnost')
        ax[0].grid()
        ax[0].legend(loc = 'best')
        
        ax[1].set_title('Validacioni skup')
        ax[1].set_xlabel('Velicina ansambla')
        ax[1].set_ylabel('Tacnost')
        ax[1].grid()
        ax[1].legend(loc = 'lower right')
        
        plt.tight_layout()
        plt.show()
```

## Random Forest

```python
n_estimators_list_rf = [1] + list(np.arange(10,210,10))
params_dict_rf = {
                  'criterion': ['gini', 'entropy'],
                  'class_weight': [None, 'balanced', 'balanced_subsample'],
                  'max_depth': list(np.arange(1,20,2)),
                  'max_features': list(np.arange(1,X_train.shape[1]))
                 }

for p in params_dict_rf.keys():
    train_ensemble_method('RF', X_train, y_train, X_test, y_test, n_estimators_list_rf, param_type=p, params_list=params_dict_rf[p])
    print('-------------------------------------------------------------------------------------------------------------------------------------------------------------')
    print('-------------------------------------------------------------------------------------------------------------------------------------------------------------')
    print('-------------------------------------------------------------------------------------------------------------------------------------------------------------')
    print('-------------------------------------------------------------------------------------------------------------------------------------------------------------')
```

Hiper-parametar "criterion":
- Na obucavajucem skupu je primetna veoma brza konvergencija ovako parametrizovanog klasifikatora, obzirom da se vec sa oko 25 stabala stize u neku stacionarnu tacnost, nezavisno od vrednosti ovog paramet
- Na validacionom skupu je primetna takodje brza konvergencija, ali sa manjom tacnoscu, sto je ocekivano zbog manjeg seta podataka; takodje, primecuje se omanja oscilatornost entropy kriterijuma u odnosu na gini kriterijum

Hiper-parametar "class_weight":
- Ovaj hiper-parametar ima 3 vrednosti: _None_, _balanced_ i _balanced subsample_
- Na obucavajucem skupu se ne primecuje znatna razlika izmedju ovih parametara, klasifikator relativno brzo dostigne dovoljnu tacnost
- Na validacionom skupu je tacnost, ponovo, manja od one na obucavajucem skupu podataka, sto je ocekivano, a pored toga javljaju se skokovi u tacnosti sa balanced i balanced_subsample parametrima 

Hiper-parametar "max_depth":
- Ovaj parametar je od izuzetne vaznosti, jer on odredjuje koliko su stabla maksimalno duboka u ansamblu
- Na obucavajucem skupu se primecuje da vece maksimalne dubine, poput 15, 17 i 19 imaju najvecu tacnost, sve ostale vrednosti imaju manju tacnost
- Na validacionom skupu su vrednosti tacnosti ocekivan, izmedju 0.75 i 0.85, i vrlo su oscilatorne

Hiper-parameter "max_features":
- Ovaj parametar predstavlja maksimalan broj odlika koje razmatra RandomForest klasifikator u cvorovima svojih stabala
- Na obucavajucem skupu je ponovo primetan veoma brza konvergencija ka vrlo viskoj tacnosti, oko 25 stabala, takodje nije primetan prevelik uticaj broja odlika na ovakve performanse
- Na validacionom skupu je primetno da samo 1 prediktora ima najvecu tacnost

## Gradient Boosting

```python
n_estimators_list_gb = [1] + list(np.arange(10,310,10))
params_dict_gb = {
                  'loss': ['deviance', 'exponential'],
                  'criterion': ['friedman_mse', 'squared_error'],
                  'max_depth': list(np.arange(1,20,2)),
                  'learning_rate': [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1]
                 }

for p in params_dict_gb.keys():
    train_ensemble_method('GB', X_train, y_train, X_test, y_test, n_estimators_list_gb, param_type=p, params_list=params_dict_gb[p])
    print('-------------------------------------------------------------------------------------------------------------------------------------------------------------')
    print('-------------------------------------------------------------------------------------------------------------------------------------------------------------')
    print('-------------------------------------------------------------------------------------------------------------------------------------------------------------')
    print('-------------------------------------------------------------------------------------------------------------------------------------------------------------')
```

Hiper-parametar "loss":
- Ovaj parametar ima dve vrednosti: _deviance_ i _exponential_
- Na obucavajucem skupu je primetna spora konvergencija ka visokoj tacnosti klasifikatora, a sa porastom broja stabala; Nesto brzi rast ima deviance parametar
- Na validacionom skupu je primetna brza konvergencija do neke ocekivane vrednosti tacnosti, oko 0.8, sa blagim oscilacijama, nezavisno od izabranog parametra

Hiper-parametar "criterion":
- Ovaj parametar ima dve vrednosti: _freidman_mse_ i _squared_error_
- Na oba skupa je primetno isto ponasanje
- Na obucavajucem skupu je primetna spora konvergencija ka visokoj tacnosti, sa porastom broja stabala
- Na validacionom skupu je primetna brza konvergencija do neke ocekivane vrednosti tacnosti, oko 0.8, sa blagim oscilacijama, nezavisno od izabranog parametra

Hiper-parametar "max_depth":
- Ovaj parametar je ponovo od izuzetne vaznosti
- Na obucavajucem skupu se primecuje da vece maksimalne dubine, poput 15, 17 i 19 imaju najvecu tacnost i znatno brze, tj. sa manje stabala dovode do te velike tacnosti, sto je ocekivano
- Na validacionom skupu su vrednosti tacnosti ocekivan, izmedju 0.75 i 0.85, a izdvaja se slucaj kada je dubina stabla 1, koji ima najvecu tacnost kada ima oko 200 stabala

Hiper-parametar "learning_rate":
- Ovaj parametar je od izuzetne vaznosti, jer on odredjuje kojom brzinom ce ovaj klasifikator stici do optimuma
- Na obucavajucem skupu se primecuje relativno brza konvergencija u pogledu tacnosti, oko 50 stabala, ali sa velikom stopom ucenje, 0.5 i 1, sto obicno nije dobro na validacionom skupu; manje vrednosti ovog parametra daju znatno losije performanse
- Na validacionom skupu se primecuje da su vrednosti stope ucenje od 0.001 pa do 0.05 optimalne i da brzo konvergiraju, ali ukoliko se krene sa velikim brojem stabala; velike vrednosti stope ucenje dovode do opadanja tacnosti sa povecavanjem broj stabala
